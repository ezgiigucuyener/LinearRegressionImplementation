```python
#We generate random values for X1 and X2 then concatenate them into X
import numpy as np

n = 100
X1 = 3 * np.random.rand(n, 1) - 1
X2 = 3 * np.random.rand(n, 1) - 1
X = np.concatenate((X1, X2), axis=1)
X
```

```python
#We assign different values to A, B, C, D to use in plane equation. We assign 2 to noise level. 
A = 2
B = 5
D = 8
C = 11
noiseLevel = 2
noise = noiseLevel * np.random.randn(n, 1)
y  = -(A*X1 + B*X2 + D) / C + noise
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#We sketch the plot using 3D points of X1, X2 and y.
ax = plt.figure(figsize=(12, 6)).gca(projection='3d')
ax.scatter(X1, X2, y, color='red', alpha=1)
ax.set(xlabel='X1', ylabel='X2', zlabel='y')

#We use prediction of linear regression of xx and yy to assign the value zz. 
m = 30 
xx, yy = np.meshgrid(np.linspace(-2, 3, m), np.linspace(-2, 3, m))
zz = -(A * xx + B * yy + D) * 1. / C # grid points on true plane
 
ax.plot_surface(xx, yy, zz, alpha=0.5)

#We plot the graph. 
plt.show()
```

```python
#We assing the input values to a matrix called X_m
X_m = np.c_[np.ones((n, 1)), X] # add x0 = 1 to each instance
X_m
```

```python
#We estimate the w_estimate value by using normal equation.
w_estimate = np.linalg.inv(X_m.T.dot(X_m)).dot(X_m.T).dot(y)
w_estimate
```

```python
#We assing new values to predict into X_second
X_second = X
X_second
```

```python
#We assing those new predicted values into another matrix called X_second_m this time. 
#We'll need these predicted data to plot the new graph. 
X_second_m = np.c_[np.ones((n, 1)), X_second] # add x0 = 1 to each instance
X_second_m
```

```python
#We predict y values using the w_estimate on the new matrix we just created. 
y_predict = X_second_m.dot(w_estimate)
y_predict
```

```python
#We plot the new graph and put the predicted points on the graph.
ax = plt.figure(figsize=(12, 6)).gca(projection='3d')
ax.plot_surface(xx, yy, zz, alpha=0.5)
ax.scatter(X1, X2, y_predict, color='green', alpha=1)
ax.set(xlabel='X1', ylabel='X2', zlabel='y')

plt.show()
```

```python
#We train our data using Linear Regression and fit our data.
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
lin_reg.predict(X_second)
```
